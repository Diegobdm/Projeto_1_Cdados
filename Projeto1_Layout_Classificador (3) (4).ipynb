### Montando um Classificador Naive-Bayes

Considerando apenas as mensagens da planilha Treinamento, ensine  seu classificador.

```python
all_words = []
all_wordsN = []
all_words_5 = []
all_words_4 = []
all_words_3 = []
all_words_2 = []
all_words_1 = []

#CLEAN UP

import re 
def cleanup(train):
    punctuation = '[!-.:?;''()]' 
    pattern = re.compile(punctuation)
    text_subbed = re.sub(pattern, '', train)
    return text_subbed

#REMOVER REPETIDOS PARA O ALL_WORDS

def remove_repetidos(lista):
    l = []
    for i in lista:
        if i not in l:
            l.append(i)
    l.sort()
    return l

#LISTA TODAS PALAVRAS

provisoria = []
for tweet in train['Treinamento']:
    splitado = tweet.split()
    for palavra in splitado:
        palavra_clean = cleanup(palavra)
        provisoria.append(palavra_clean)
all_words = remove_repetidos(provisoria) #all_words eh uma lista sem palavras repetidas agora

#NOTA 5
        
rate_5 = train['CLASSIFICACAO'] == 5
tweets_5 = train.loc[rate_5, :]
for tweet in tweets_5['Treinamento']:
    splitado = tweet.split()
    for palavra in splitado:
        palavra_clean = cleanup(palavra)
        all_words_5.append(palavra_clean)

#NOTA 4
       
rate_4 = train['CLASSIFICACAO'] == 4
tweets_4 = train.loc[rate_4, :]
for tweet in tweets_4['Treinamento']:
    splitado = tweet.split()
    for palavra in splitado:
        palavra_clean = cleanup(palavra)
        all_words_4.append(palavra_clean)

#NOTA 3
       
rate_3 = train['CLASSIFICACAO'] == 3
tweets_3 = train.loc[rate_3, :]
for tweet in tweets_3['Treinamento']:
    splitado = tweet.split()
    for palavra in splitado:
        palavra_clean = cleanup(palavra)
        all_words_3.append(palavra_clean)

#NOTA 2
       
rate_2 = train['CLASSIFICACAO'] == 2
tweets_2 = train.loc[rate_2, :]
for tweet in tweets_2['Treinamento']:
    splitado = tweet.split()
    for palavra in splitado:
        palavra_clean = cleanup(palavra)
        all_words_2.append(palavra_clean)

#NOTA 1
       
rate_1 = train['CLASSIFICACAO'] == 1
tweets_1 = train.loc[rate_1, :]
for tweet in tweets_1['Treinamento']:
    splitado = tweet.split()
    for palavra in splitado:
        palavra_clean = cleanup(palavra)
        all_words_1.append(palavra_clean)
```

```python
def nota(tweet):
    tweet_words_uncleaned = tweet.split()
    tweet_words = [] #palavras limpas do tweet
    for palavra in tweet_words_uncleaned:
        tweet_words.append(cleanup(palavra))
    P5 = 1
    P4 = 1
    P3 = 1
    P2 = 1
    P1 = 1
    for word in tweet_words:
        frequencia_em_5 = all_words_5.count(word)
        P5_word = (frequencia_em_5+1)/(len(all_words_5)+len(all_words))
        P5 *= P5_word
        frequencia_em_4 = all_words_4.count(word)
        P4_word = (frequencia_em_4+1)/(len(all_words_4)+len(all_words))
        P4 *= P4_word
        frequencia_em_3 = all_words_3.count(word)
        P3_word = (frequencia_em_3+1)/(len(all_words_3)+len(all_words))
        P3 *= P3_word
        frequencia_em_2 = all_words_2.count(word)
        P2_word = (frequencia_em_2+1)/(len(all_words_2)+len(all_words))
        P2 *= P2_word
        frequencia_em_1 = all_words_1.count(word)
        P1_word = (frequencia_em_1+1)/(len(all_words_1)+len(all_words))
        P1 *= P1_word
    lista_comparacao = [P1, P2, P3, P4, P5]
    maior = max(lista_comparacao)
    if maior == P5:
        return 5
    elif maior == P4:
        return 4
    elif maior == P3:
        return 3
    elif maior == P2:
        return 2
    elif maior == P1:
        return 1
```

```python
j = 0
for tweet in test['Teste']:
    rate = nota(tweet)
    test['CLASSIFICACAO_MACHINE'][j] = rate
    j+=1
test.head(15)
```
